- Following is an interactive-style exploratory data analysis of the SUPPORT-II dataset. It is designed to be run from the command line (as well) and will produce both a text summary and a plot.

```python
import sys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import our custom functions from the 'eda_report.py' module.

try:
    sys.path.append('../src')
    from eda_report import generate_full_report, missing_summary, load_data
except ImportError:
    print("Error: Could not import from 'src/eda_report.py'.")
    print("Please ensure you are running this script from the 'notebooks' directory,")
    print("or that the project structure is correct.")
    sys.exit(1)
```

```python
# Configure plot style for better a e s t h e t i c s.

sns.set_theme(style="whitegrid")
print("Setup complete. Functions imported.\n")
```

```python
# Generate Command-Line EDA Report
# Here, we call the main reporting function to get a high-level,
# text-based overview of the entire dataset.

print("Generating full text-based EDA report...")
generate_full_report()
```

```python
# Interactive Analysis: Visualizing Missing Data 
# Now we generate the visual component.
# get an intuitive sense of the patterns of missingness in the data.

print("\nGenerating visual heatmap for missing data...")

# Load the data using our helper function
df = load_data()

if df is not None:
    # Generate the missingness table and display the heatmap.
    # toggled plot_heatmap=True to activate the visualization
    missing_df = missing_summary(df, plot_heatmap=True)
    
    print("\nSummary of Features with Missing Data:")
    print(missing_df.to_string())
    
    print("\nShowing missing data heatmap plot...")
    plt.show()

print("\nEDA script finished.")
```
